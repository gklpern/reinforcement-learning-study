```python
import gymnasium as gym
import time
```

```python
env_name = 'MountainCar-v0'
env = gym.make(env_name,render_mode='human')
```

##### Observation - ilk gözlem   ,   info   -   gerekli bilgiler.

```python
observation, info = env.reset(seed=42)
```

```python
print("Aksiyon Uzayı:", env.action_space)
```

```python
if isinstance(env.action_space, gym.spaces.Discrete):
    print("Olası Aksiyonlar:", [i for i in range(env.action_space.n)])
```

```python
def simple_agent(observation):

    # get observation
    position,velocity=observation

    # action
    if -0.1 < position < 0.4:
        action = 2

    elif velocity < 0 and position < -0.2:
        action = 0

    else:
        action = 1


    return action
```

```python
for step in range(200):
    action = env.action_space.sample()


    action = simple_agent(observation)

    observation, reward, terminated, truncated, info = env.step(action)

    print(f'Velocity: {observation[1]}')
    print(f'Position: {observation[0]}')

    time.sleep(0.1)
               
env.close()
```
